```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Generar datos para la distribución normal

media = 70 # Media
desv_std = 4 # Desviación estándar
num_datos = 500

# Generamos los datos de manera aleatoria y los ponemos en una lista (array)
datos = np.random.normal(media, desv_std, num_datos)
datos[0:50] # Mostrar los primeros 50
```

```python
# Crear un rango de valores para crear una densidad
x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, num_datos)

# Calcular la PDF de la distribución normal
pdf = 1 / (desv_std * np.sqrt(2 * np.pi)) * np.exp(-(x - media)**2 / (2 * desv_std **2))
```

```python
pdf.size, num_datos
```

```python
# Graficar histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, color='b') # Histograma de los datos
plt.title('Distribución Normal ')
plt.xlabel('Peso en Kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
# Visualizar la PDF
plt.figure(figsize=(6, 4))
plt.plot(x, pdf, label='Normal:(media:{}; ds:{})'.format(np.round(media,2), np.round(desv_std, 2)))
plt.title('Función de Densidad de Probabilidad (PDF) - Distrubución Normal')
plt.xlabel('Peso en Kg de los estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Graficar histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b') # Histograma de los datos
plt.plot( x, pdf, 'k', linewidth=2, )
plt.title('Distribución Normal ')
plt.xlabel('Peso en Kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
# Parámetros
media = 70
desv_std = 4
total_estudiantes = 500

# Límites de interés
peso_min = 68
peso_max = 72

# Dominio para la gráfica
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

# Calcular la probabilidad de estar entre 68 y 72 kg
prob_entre = norm.cdf(peso_max, media, desv_std) - norm.cdf(peso_min, media, desv_std)
estudiantes_entre = prob_entre * total_estudiantes

# Graficar la distribución y el área de interés
plt.figure(figsize=(6, 4))
plt.plot(x, y, 'k', linewidth=2, label='Distribución Normal')
plt.fill_between(x, y, where=(x >= peso_min) & (x <= peso_max), color='skyblue', alpha=0.5,
                 label='Entre 68 y 72 kg')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribución Normal: Estudiantes entre 68 y 72 kg')
plt.legend()
plt.grid(True)
plt.show()

print("Probabilidad entre 68 y 72 kg:", prob_entre)
print("Número aproximado de estudiantes entre 68 y 72 kg:", estudiantes_entre)
```
